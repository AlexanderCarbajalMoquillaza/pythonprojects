```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fifa.csv')
```

```python
#mostrar filas y columnas
df.shape
```

```python
#leer la tabla
df
```

```python
#visualizar los tipos de datos y datos perdidos
df.info()
```

```python
df.describe()
```

```python
#limpiando datos de columnas: possession team1, possession team2
df['possession team1'] = df['possession team1'].str.replace('%','')
df['possession team2'] = df['possession team2'].str.replace('%','')
```

```python
#transformando el campo 'date' de cadena(object) a datetime64
df['date'] = df['date'].astype('datetime64')
```

```python
df.info()
```

```python
#contar veces que aparece un equipo en 'team1'
part_jugados_t1 = df['team1'].value_counts()
print(part_jugados_t1)

#contar veces que aparece un equipo en 'team2'
part_jugados_t2 = df['team2'].value_counts()
print(part_jugados_t2)
```

```python
#suma de partidos jugados de Team 1 y Team 2
part_jugados_tot = part_jugados_t1 + part_jugados_t2 

#ordenar partidos jugados por país
part_jugados_tot_ord = part_jugados_tot.sort_values(ascending=False)
print(part_jugados_tot_ord)
```

```python
g1 = part_jugados_tot_ord 
g1.plot(kind='bar', figsize=(15,8), color='#690A2E',
       edgecolor='#EDEAE8',
       width=0.5,
       lw=2)

plt.title('Partidos jugados por país - Mundial Qatar 2022')
plt.xlabel('Países')
plt.ylabel('Partidos jugados')
```

```python
#la información inicial fue brindada por 'team1' y 'team2'. 
#De 'team1' se está identificando por fecha la cantidad de tarjetas amarillas
g2 = df.groupby(by='date')['yellow cards team1'].sum()
g2.plot(figsize=(15,10),
        color ='#C4102D', #color de linea
       lw = 3, #grosor de linea
       marker = 'o', #forma de puntero
       markerfacecolor='#EDEAE8', #color de puntero
       markersize=10
       )
```

### Identificar cantidad de amarillas por equipo

```python
#Para visualizar mejor, separo las columnas a necesitar en otra variable

df_amarillas = df[['team1','yellow cards team1','team2','yellow cards team2']]
df_amarillas
```

```python
#contando amarillas por pais pero de 'team1'
ta1 = df_amarillas.groupby('team1')['yellow cards team1'].sum()
ta1
```

```python
#contando amarillas por pais pero de 'team2'
ta2 = df_amarillas.groupby('team2')['yellow cards team2'].sum()
ta2
```

```python
#sumar amarillas contadas de 'team1' y 'team2' del mismo país para que muestre el total de amarillas por país en el mundial
tot_amarillas = ta1 + ta2

#ordenando de país con más tarjetas amarillas a país con menos tarjetas amarillas
tot_amarillas_ord= tot_amarillas.sort_values(ascending=False)
print(tot_amarillas_ord)
```

```python
g3 = tot_amarillas
g3.plot(kind='bar', figsize=(15,8), color='#690A2E',
       edgecolor='#EDEAE8',
       width=0.5,
       lw=2)

plt.title('Total de tarjetas amarillas por país - Mundial Qatar 2022')
plt.xlabel('Países')
plt.ylabel('Cantidad de tarjetas amarillas')
```

### Identificar cantidad de rojas por equipo

```python
#Para visualizar mejor, separo las columnas a necesitar en otra variable

df_rojas = df[['team1','red cards team1','team2','red cards team2']]
df_rojas
```

```python
tr1 = df_rojas.groupby('team1')['red cards team1'].sum()
tr1
```

```python
#contando tarjetas rojas por pais pero de 'team2'
tr2 = df_rojas.groupby('team2')['red cards team2'].sum()
tr2
```

```python
#sumar rojas contadas de 'team1' y 'team2' del mismo país para que muestre el total de rojas por país en el mundial
tot_rojas = tr1 + tr2

#ordenando de país con más tarjetas rojas a país con menos tarjetas rojas
tot_rojas_ord= tot_rojas.sort_values(ascending=False)
print(tot_rojas_ord)
```

```python
g4 = tot_rojas
g4.plot(kind='bar', figsize=(15,5), color='#690A2E',
       edgecolor='#EDEAE8',
       width=0.5,
       lw=2
       )

plt.title('Total de tarjetas rojas por país - Mundial Qatar 2022')
plt.xlabel('Países')
plt.ylabel('Cantidad de tarjetas rojas')
```

### Cantidad de amarillas por fecha para Argentina

```python
#contando amarillas de Argentina pero de 'team1'
taa1 = df[df['team1']=='ARGENTINA'].groupby('date')['yellow cards team1'].sum()
taa1
```

```python
taa2 = df[df['team2']=='ARGENTINA'].groupby('date')['yellow cards team2'].sum()
taa2
```

```python
tarj_amarillas_arg = pd.concat([taa1,taa2],axis=0)
tarj_amarillas_arg 
```

```python
g5 = tarj_amarillas_arg
g5.plot(figsize=(15,8),
        color ='#C4102D', #color de linea
       lw = 3, #grosor de linea
       marker = 'o', #forma de puntero
       markerfacecolor='#EDEAE8', #color de puntero
       markersize=10,
        grid=True
       )

plt.title('Total de tarjetas amarillas para Argentina durante el Mundial Qatar 2022')
plt.xlabel('Fechas')
plt.ylabel('Cantidad de tarjetas amarillas')
```
