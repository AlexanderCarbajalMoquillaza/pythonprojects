```python
#imputar cuando falten datos o missing value
#imputar los 177 valores faltantes de Age con la mediana
#tecnicas para imputar valores faltante: imputar por la media, mediana, media condicional, mediana condicional

#por la mediana simple (la mas utilizada)
mediana_edad = data['Age'].median()
print(mediana_edad)
data['Age'] = data['Age'].fillna(mediana_edad)
```

```python
#Analisis exploratorio con visualizacion de datos
# Visualizar superviviencia por genero
# crear uan columna 'died' (complemento de survived)

data['died'] = 1 - data['Survived']
data[['Survived','died']].head(6)
```

```python
#precio promedio de la tarifa por clase
g5= data.groupby('Pclass').mean()['Fare']
g5.plot(kind='bar', figsize=(10,5))
```

## Modelo de Machine Learning: Problema Clasificacion
### Regresion Logistica

```python
#importar libreria sklearn
from sklearn.model_selection import train_test_split
```

```python
#partir en entrenamiento y test
y = data['Survived']
x = data[['Pclass','Age','SibSp','Parch','Fare','gender']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size= 0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#entreno modelo
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
predict = lr.predict(x_test)
predict
```
